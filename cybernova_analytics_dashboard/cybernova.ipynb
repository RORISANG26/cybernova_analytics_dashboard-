```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Config
num_rows = 6000

countries = {
    "Botswana": "Gaborone",
    "South Africa": "Johannesburg",
    "Kenya": "Nairobi",
    "Nigeria": "Lagos",
    "UK": "London",
    "USA": "New York",
    "India": "Mumbai"
}

pages = ["Home", "Services", "About", "Contact", "Blog", "Pricing"]
services = ["AI Security", "Cloud Security", "Risk Assessment", "Data Protection"]
devices = ["Mobile", "Desktop"]
traffic_sources = ["Organic", "Paid Ads", "Social Media", "Direct"]

data = []
start_date = datetime(2024, 1, 1)

for i in range(num_rows):
    user_id = random.randint(1, 1200)
    session_id = random.randint(10000, 99999)

    timestamp = start_date + timedelta(
        days=random.randint(0, 180),
        hours=random.randint(0, 23),
        minutes=random.randint(0, 59)
    )

    # Geography
    country = random.choice(list(countries.keys()))
    region = countries[country]

    # Traffic source influences behavior
    traffic_source = random.choice(traffic_sources)

    # Page bias
    if traffic_source == "Paid Ads":
        page = random.choices(pages, weights=[10, 40, 5, 10, 5, 30])[0]
    else:
        page = random.choice(pages)

    service = random.choice(services)

    # Device behavior
    device = random.choice(devices)

    if device == "Mobile":
        time_spent = np.random.randint(5, 180)
    else:
        time_spent = np.random.randint(20, 400)

    # Action logic (conversion bias)
    if traffic_source == "Paid Ads":
        action = random.choices(
            ["Page View", "Click", "Scroll", "Demo Request", "Event Registration"],
            weights=[35, 20, 15, 20, 10]
        )[0]
    else:
        action = random.choices(
            ["Page View", "Click", "Scroll", "Demo Request", "Event Registration"],
            weights=[55, 20, 15, 7, 3]
        )[0]

    converted = 1 if action in ["Demo Request", "Event Registration"] else 0

    data.append([
        user_id, session_id, timestamp, country, region,
        page, action, time_spent, device, traffic_source,
        service, converted
    ])

df = pd.DataFrame(data, columns=[
    "user_id", "session_id", "timestamp", "country", "region",
    "page", "action", "time_spent", "device", "traffic_source",
    "service", "converted"
])

# ---- FEATURE ENGINEERING ---- #

# Engagement level
def engagement_level(time):
    if time > 250:
        return "High"
    elif time > 120:
        return "Medium"
    else:
        return "Low"

df["engagement_level"] = df["time_spent"].apply(engagement_level)

# User type (simulate returning users)
df["user_type"] = np.where(df["user_id"] % 3 == 0, "Returning", "New")

# Save
df.to_csv("cybernova_dataset.csv", index=False)

print("✅ Smart dataset generated successfully!")
```
